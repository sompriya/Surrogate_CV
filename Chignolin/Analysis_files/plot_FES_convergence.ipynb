```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l1 = np.loadtxt('TDA_lasso_run1_barr30/deltaF.dat')
l2 = np.loadtxt('TDA_lasso_run2_barr30/deltaF.dat')
l3 = np.loadtxt('TDA_lasso_run3_barr30/deltaF.dat')

print(l1.shape)
print(l2.shape)
print(l3.shape)

#just to make the array sizes same. should be deleted later
l2 = l2[:200]
l3 = l3[:200]

print(l2.shape)
print(l3.shape)
```

```python
Fmean = np.mean(np.vstack((l1,l2,l3)),axis=0)
print(Fmean.shape)

Fstd = np.std(np.vstack((l1,l2,l3)),axis=0)
```

```python
xx = np.arange(len(Fmean))*0.01

#plt.plot(xx,Fmean)

plt.ylim(-15,15)
plt.xlim(0.05,2)

plt.fill_between(xx,Fmean-Fstd,Fmean+Fstd,color='xkcd:lightgreen')
plt.plot(xx,l1,color='g',lw=1)
plt.plot(xx,l2,color='g',lw=1)
plt.plot(xx,l3,color='g',lw=1)

reference = 4.1855*0.9 #kJ/mol

plt.plot(xx,np.ones(len(xx))*reference,'k--',label='Reference')

plt.xlabel('Simulation Time ($\mu$s)',fontsize=16)
plt.ylabel('$\Delta F_U$ (kJ/mol)',fontsize=16)

plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.title('Surrogate Model CV',fontsize=16)

plt.savefig('TDA_lasso_conv_chignolin.png',dpi=300,bbo
```
